```python
import gmaps
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import scipy.stats as stats
from keys import gkey
```

```python
gmaps.configure(api_key=gkey)
happdf = pd.read_csv('../Data/capitals.csv')
happdf
```

```python
# use capitals.csv
# obtain geocode for country, city
# increase weight in heatmap (read documentation)
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json?'
lats = []
lngs = []
ranks = []
for cntry, rank, cap  in zip(happdf['Country'], happdf['Ladder'], happdf['Capital City']):
    response = requests.get(url, params ={'key':gkey, 'address': f"{cntry},{cap}"}).json()
    lats.append(response['results'][0]['geometry']['location']['lat'])
    lngs.append(response['results'][0]['geometry']['location']['lng'])
    ranks.append(rank)
```

```python
locations = pd.DataFrame({'lats':lats,'lngs':lngs, 'ranks':ranks})
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations[['lats','lngs']], 
                                 dissipating=False, max_intensity=10,weights=(1/locations['ranks'])*100,
                                 point_radius=5)
fig.add_layer(heat_layer)
fig
```

```python
heat_layer = gmaps.heatmap_layer(locations[['lats','lngs']], 
                                 dissipating=False, max_intensity=5,weights=(locations['ranks']/156)*100,
                                 point_radius=5)
fig.add_layer(heat_layer)
fig
```
